```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
import statsmodels.api as sm
```

```python
#Loading datasets
data1 = pd.read_csv('historical_data.csv')
data2 = pd.read_csv('fear_greed_index.csv')
```

```python
# Handle missing values and ensure correct data types 
data1 = data1.dropna()  
data1 = data1.convert_dtypes()  

data2 = data2.dropna()
data2 = data2.convert_dtypes()  
```

```python
print(data1.columns)
print(data2.columns)
```

```python
data1.rename(columns={'Timestamp IST': 'date'}, inplace=True)
```

```python
print(data1.columns)
```

```python
# Convert the 'date' column to datetime format, coercing invalid entries to NaT
data1['date'] = pd.to_datetime(data1['date'], errors='coerce')

# Drop rows where 'date' is NaT (invalid dates)
data1 = data1.dropna(subset=['date'])

# Format the 'date' column to keep only the date in DD/MM/YY format
data1['date'] = data1['date'].dt.strftime('%d/%m/%y')
```

```python
data1.head()
```

```python
# Ensure the datetime format for both datasets
data1['date'] = pd.to_datetime(data1['date'])
data2['date'] = pd.to_datetime(data2['date'])

# Merge the datasets 
merged_data = pd.merge(data1, data2, on='date', how='inner')  
```

```python
# Save the merged dataset to a new CSV file
merged_data.to_csv('merged_data.csv', index=False)
```
